## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

import warnings
warnings.filterwarnings("ignore")

#  Import and read the charity_data.csv.
import pandas as pd
df_application = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
df_application.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
df_applications = df_application.drop(columns = ["EIN", "NAME"])
df_applications.head()
```

```python
# Determine the number of unique values in each column.
df_applications.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
count_apps =df_applications.APPLICATION_TYPE.value_counts()
count_apps
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = list(count_apps[count_apps < 500].index)

# Replace in dataframe
for app in application_types_to_replace:
    df_applications['APPLICATION_TYPE'] = df_applications['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
df_applications['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
count_class =df_applications.CLASSIFICATION.value_counts()
count_class
```

```python
#view classification counts > 1
count_class_gtone = count_class.loc[count_class > 1]
count_class_gtone
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
#  YOUR CODE GOES HERE
classifications_to_replace = list(count_class[count_class < 800])
# Replace in dataframe
for cls in classifications_to_replace:
    df_applications['CLASSIFICATION'] = df_applications['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
df_applications['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
df_app_num = pd.get_dummies(df_applications)
df_app_num.head()
```

```python
from os import X_OK
# Split our preprocessed data into our features and target arrays
X= df_app_num.drop(['IS_SUCCESSFUL'], axis = 'columns').values
y = df_app_num['IS_SUCCESSFUL'].values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# define the model - deep neural net
nn_model = tf.keras.models.Sequential()

number_input_features = len(X_train[0])
number_nodes_1 = 90
number_nodes_2 = 30
number_nodes_output = 1


# add first hidden layer
nn_model.add(
    tf.keras.layers.Dense(units=number_nodes_1, input_dim=number_input_features, activation="relu")
)

# add second hidden layer
nn_model.add(tf.keras.layers.Dense(units = number_nodes_2, input_dim = number_input_features, activation = "tanh"))

# add output layer

nn_model.add((tf.keras.layers.Dense(units = number_nodes_output, input_dim = number_input_features, activation = "sigmoid")))

# Check the structure of the model
nn_model.summary()


```

```python
# Compile the model
nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn_model.fit(X_train_scaled, y_train, epochs=50)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
